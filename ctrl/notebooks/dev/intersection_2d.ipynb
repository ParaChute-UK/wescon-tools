```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = 0
y1 = 0
u1 = 1
v1 = 1

x2 = 3
y2 = 1
u2 = 0.5
v2 = 1
```

```python
M = np.matrix([[u1, -u2], [v1, -v2]])
b = np.array([x2 - x1, y2 - y1])
```

```python
f = 7
plt.plot(x1, y1, 'or')
plt.plot([x1, x1 + u1 * f], [y1, y1 + v1 * f], '-r')

plt.plot(x2, y2, 'ob')
plt.plot([x2, x2 + u2 * f], [y2, y2 + v2 * f], '-b')
# plt.plot([x2, y2], [u2 * 5, v2 * 5])

a1 = (M.I @ b)[0, 0]
plt.plot(x1 + a1 * u1, y1 + a1 * v1, marker='o', color='purple')
```

```python
def plot_intersection(x1, y1, u1, v1, x2, y2, u2, v2, f=7):
    plt.plot(x1, y1, 'or')
    plt.plot([x1, x1 + u1 * f], [y1, y1 + v1 * f], '-r')

    plt.plot(x2, y2, 'ob')
    plt.plot([x2, x2 + u2 * f], [y2, y2 + v2 * f], '-b')
    # plt.plot([x2, y2], [u2 * 5, v2 * 5])

    M = np.matrix([[u1, -u2], [v1, -v2]])
    b = np.array([x2 - x1, y2 - y1])
    a1 = (M.I @ b)[0, 0]
    plt.plot(x1 + a1 * u1, y1 + a1 * v1, marker='o', color='purple')
```

```python
plot_intersection(0, 0, 1, 1, 3, 1, 0.3, 1)
```
